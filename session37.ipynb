```python
def read_binary_file(filename, chunk_size=10):
    with open(filename, mode="rb") as instream:
        while (chunk := instream.read(chunk_size)):
            yield chunk


for block in read_binary_file("/bin/ls"):
    print(block)
```

```python
with open("inventory.csv") as inv:
    headers = [field.strip('"') for field in inv.readline().rstrip().split(",")]
    inventory = dict().fromkeys(headers, [])
    print(id(inventory["product"]), id(inventory["quantity"]))
    print(inventory)
    for line in inv:
        product, quantity, price = line.rstrip().split(",")
        inventory[headers[0]].append(product.strip('"'))
        inventory[headers[1]].append(int(quantity))
        inventory[headers[2]].append(int(price))

    print(inventory)

```
